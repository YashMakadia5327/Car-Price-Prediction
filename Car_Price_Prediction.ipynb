```python
import pandas as pd
```

```python
car = pd.read_csv('cardekho.csv')
```

```python
car.head()
```

```python
car.dropna(inplace=True)
```

```python
car.isnull().sum()
```

```python
car=car.drop('seller_type',axis=1)
```

```python
car.head()
```

```python
car.describe()
```

```python
car = car[car['selling_price']<6e6].reset_index(drop=True)
```

```python
car.head()
```

```python
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car.head()
```

```python
car['seats']= car['seats'].astype(int)
car['mileage(km/ltr/kg)']= car['mileage(km/ltr/kg)'].astype(int)
car['engine']= car['engine'].astype(int)
```

```python
car.head()
```

```python
car['owner'] = car['owner'].str.replace('Owner','')
```

```python
car.head()
```

```python
car.to_csv("clean_car.csv")
```

```python
car.info()
```

Model

```python
x = car.drop('selling_price',axis=1)
y = car['selling_price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','fuel','transmission','owner','max_power']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','fuel','transmission','owner','max_power']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_pred,y_test)
```

```python
score=[]
for i in range(7907):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred = pipe.predict(x_test)
    score.append(r2_score(y_pred,y_test))
```

```python
import numpy as np
np.argmax(score)
```

```python
score[np.argmax(score)]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=np.argmax(score))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
r2_score(y_pred,y_test)
```
